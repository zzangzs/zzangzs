#### 1.4. Finding the angle between two vectors
Find the angle between two vectors using matrix operations.  <br>
A = [0, 1, 0]$^\text{T}$ <br>
B = [2, 0, 0]$^\text{T}$ <br>

```python
N = 10
A = np.random.rand(N, N) # Square matrix
B = np.random.rand(N).T # N-length vector
##1.1
def matmul1(A,B):
    C = np.zeros(A.shape[0]).T
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            C[i]+=A[i,j]*B[j]
    return C
    
##1.2
def matmul2(A,B):
    C = np.zeros(A.shape[0]).T
    for i in range(A.shape[0]):
        C[i] = np.dot(A[i,:],B)
    return C

##1.3
print(matmul1(A,B))
print(matmul2(A,B))
print(A@B)

##1.4
A = np.array([0,1,0]).T
B = np.array([2,0,0]).T

angle = np.arccos(np.dot(A,B) / (np.linalg.norm(A)*np.linalg.norm(B)))
print(angle)
```

### Exercise 2: Fourier Series as Matrix Multiplication
Recall that a Fourier Series is an approximation of a periodic function as a summation of harmonic sinusoids. 

```python
# Code here is provided as an example, no need to edit.

t = np.linspace(0, 10, 1000) #sampling freq = 100Hz
f_ish = np.zeros(1000)
K = np.arange(1, 21, 2) # array of = 2k-1
for k in K:
    sinusoid = np.sin(2*np.pi*k*t)
    coefficient = 4/np.pi/k
    f_ish += coefficient*sinusoid
    if (k == 1 or k == 3 or k == 5):
        plt.plot(t, coefficient*sinusoid, label='$sin(%i\pi t)$' %(2*k))
f = 2*((2*t+1)%2//1-0.5)
plt.plot(t, f, label='function')
plt.plot(t, f_ish, label='approximation')
plt.title('')
plt.xlim([0, 1])
plt.ylim([-1.5,1.5])
plt.legend()
plt.show
plt.figure()
```

If you look at the code in the cell above, you can see that I gave you a time vector t and square wave function f (which is a shifted and scaled version of the square wave you wrote before).
In this exercise, I want to find the Fourier series coefficients of the square wave function. I provided the vector of different values of $2k-1$ called `K`, which you will use to generate sinusoids. 

How do you get the coefficients of a given sinusoid? You can take the inner product (dot product in our case) between the signal and the sinusoid. 
#### Task: Plot the Fourier coefficients that you just calculated (against the linear frequency f) and compare it with Exercise 6 from last week's assignment. 

```python
K = np.arange(1, 21, 1) 
f = 2*((2*t+1)%2//1-0.5) # Square wave function 
F = np.zeros(K.shape[0]) # array of coefficients you calculated
i = 0
for k in K:
    sinusoid =  np.sin(2*np.pi*k*t)# sinusoid
    F[i] = np.dot(f,sinusoid)# dot product
    i+=1
plt.stem(K, F/500, use_line_collection=True)
```

Consider the circuit above: $R$  = 2 $\Omega$, $L$ = 5$H$, $C$ = 10$F$, $V_S$ = 2sin(2t)

Find the steady state voltage for the resistor using phasors. Plot the input and steady-state output on the same plot. Label all the plots accordignly.

```python
t = np.linspace(0,10,1000)
V = 2 * np.sin(2*t)

#Solve for and plot the steady state response here
ZR = 2
ZC = 1/20j
ZL = 10j

VR_Phasor = ZR/(ZR+ZC+ZL)*2
VR = np.absolute(VR_Phasor)*np.sin(2*t+np.angle(VR_Phasor))
plt.plot(t,V,label='input')
plt.plot(t,VR,label='output')
plt.xlabel('t')
plt.title('plot of $V_s$ and $V_{ss}$ for the resistor')
plt.legend(loc='best')
plt.show()
```
